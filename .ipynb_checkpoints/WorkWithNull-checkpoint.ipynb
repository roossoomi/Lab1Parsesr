```python
import pandas as pd
import json
import re

data = pd.read_json('parametrs.json', orient='index')
numeric_cols = ['grams_protein', 'percent_protein', 'grams_carb', 'percent_carb', 'grams_fat',
               'percent_fat', 'calories', 'low_carb', 'medium_carb', 'high_carb', 'number_of_steps']
data[numeric_cols] = data[numeric_cols].astype(float)
```

```python
for i in range (0, 13):
    data.iloc[i, i+1] = pd.NA
for i in range(0,5):
    data.iloc[i, 1] = pd.NA
data.head()
```

```python
data.author = data.author.fillna(data.author.mode()[0])
data.head()
```

```python
print(data.sort_values(['grams_fat', 'grams_carb'], ascending=[False, False]))
```

```python
print(data)
```

```python
data.ffill()
print(data)
```

```python
url_base = 'https://alexfitness.ru/fitness_guide/norma_upotrebleniya_kalorij'
tables = pd.read_html(url_base)
tables[1]
```

```python
text = tables[1][1][2]
pattern = r"(\d+)\–(\d+)" 
match = re.search(pattern, text)
lower_value = int(match.group(1))
upper_value = int(match.group(2)) 
data['percentageOfDailyAllowance'] = (data['calories'] / ((lower_value + upper_value) / 2) * 100).round(2)
print(data)
```
